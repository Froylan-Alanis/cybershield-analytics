```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet(
    '../data/clean/cybersecurity_attacks_clean.parquet'
)
```

```python
df.head()
```

```python
attack_counts = df['attack_type'].value_counts()

attack_counts
```

```python
plt.figure(figsize=(10,6))

attack_counts.plot(kind='bar')

plt.title('Distribución de Tipos de Ataque')
plt.xlabel('Tipo de Ataque')
plt.ylabel('Cantidad')

plt.xticks(rotation=45)

plt.show()
```

```python
severity_counts = df['severity_level'].value_counts()

severity_counts
```

```python
plt.figure(figsize=(8,5))

severity_counts.plot(kind='bar')

plt.title('Distribución de Severidad')
plt.xlabel('Nivel de Severidad')
plt.ylabel('Cantidad')

plt.xticks(rotation=0)

plt.show()
```

```python
state_counts = df['mexico_state'].value_counts()

state_counts.head(10)
```

```python
plt.figure(figsize=(12,6))

state_counts.head(10).plot(kind='bar')

plt.title('Top Estados Mexicanos con Eventos')
plt.xlabel('Estado')
plt.ylabel('Cantidad de Eventos')

plt.xticks(rotation=45)

plt.show()
```
