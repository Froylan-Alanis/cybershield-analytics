```python
# Estrategia de tratamiento de valores nulos
"Las columnas relacionadas con alertas, malware y proxies contienen valores nulos que pueden representar ausencia de eventos de seguridad, por lo que inicialmente no serán eliminadas."
```

```python
df_clean = df.copy()
```

```python
df_clean.shape
```

```python
india_regions = (
    df_clean['geo_location_data']
    .str.split(',')
    .str[-1]
    .str.strip()
)
```

```python
india_regions.unique()[:30]
```

```python
india_regions.value_counts()
```

```python
mexico_states = [
    'Ciudad de México',
    'Jalisco',
    'Nuevo León',
    'Puebla',
    'Querétaro',
    'Yucatán',
    'Chihuahua',
    'Guanajuato',
    'Veracruz',
    'Baja California',
    'Sonora',
    'Coahuila',
    'Tamaulipas',
    'Michoacán',
    'Oaxaca',
    'Sinaloa',
    'Durango',
    'San Luis Potosí',
    'Aguascalientes',
    'Zacatecas'
]
```

```python
india_unique = india_regions.unique()

mapping_dict = {}

for i, region in enumerate(india_unique):
    mapping_dict[region] = mexico_states[i % len(mexico_states)]
```

```python
list(mapping_dict.items())[:20]
```

```python
df_clean['mexico_state'] = india_regions.map(mapping_dict)
```

```python
df_clean[['geo_location_data', 'mexico_state']].head(20)
```

```python
df_clean['mexico_state'].isnull().sum()
```

```python
# Transformación Geográfica India → México

'Se implementó una estrategia de regionalización para contextualizar el dataset original de India hacia entidades geográficas mexicanas, permitiendo análisis espaciales y dashboards orientados al contexto nacional.'
```

```python
mexico_coordinates = {
    'Ciudad de México': (19.4326, -99.1332),
    'Jalisco': (20.6597, -103.3496),
    'Nuevo León': (25.6866, -100.3161),
    'Puebla': (19.0414, -98.2063),
    'Querétaro': (20.5888, -100.3899),
    'Yucatán': (20.9674, -89.5926),
    'Chihuahua': (28.6329, -106.0691),
    'Guanajuato': (21.0190, -101.2574),
    'Veracruz': (19.1738, -96.1342),
    'Baja California': (32.5149, -117.0382),
    'Sonora': (29.0729, -110.9559),
    'Coahuila': (25.4267, -100.9954),
    'Tamaulipas': (23.7369, -99.1411),
    'Michoacán': (19.7008, -101.1844),
    'Oaxaca': (17.0732, -96.7266),
    'Sinaloa': (24.8091, -107.3940),
    'Durango': (24.0277, -104.6532),
    'San Luis Potosí': (22.1565, -100.9855),
    'Aguascalientes': (21.8853, -102.2916),
    'Zacatecas': (22.7709, -102.5832)
}
```

```python
df_clean['latitude'] = df_clean['mexico_state'].map(
    lambda x: mexico_coordinates[x][0]
)

df_clean['longitude'] = df_clean['mexico_state'].map(
    lambda x: mexico_coordinates[x][1]
)
```

```python
df_clean[['mexico_state', 'latitude', 'longitude']].head(20)
```

```python
df_clean[['latitude', 'longitude']].isnull().sum()
```

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
geometry = [
    Point(xy)
    for xy in zip(df_clean['longitude'], df_clean['latitude'])
]
```

```python
gdf = gpd.GeoDataFrame(
    df_clean,
    geometry=geometry,
    crs='EPSG:4326'
)
```
